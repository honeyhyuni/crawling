# 교촌 치킨 정적 웹 크롤링

```python
from bs4 import BeautifulSoup
import urllib.request
import pandas as pd
import datetime
```

```python
def gyochon_store(result):
    dic = {1: 26, 2: 17, 3: 9, 4: 11, 5: 6, 6: 6, 7: 6, 8: 17, 9: 45, 10: 19, 11: 16, 12: 17, 13: 16,
          14: 23, 15: 25, 16: 23, 17: 3}  # sido2 를 돌기 위한 딕셔너리
    for page in range(1,18):
        for page2 in range(1, dic[page]):
            gyochon_url = 'https://kyochon.com/shop/domestic.asp?sido1=%d&sido2=%d&txtsearch=' %(page, page2)
            print(gyochon_url)
            html = urllib.request.urlopen(gyochon_url)
        
            soupGyochon = BeautifulSoup(html, 'html.parser')
            store_gyochon = soupGyochon.find(class_ = 'shopSchList')
            for store in store_gyochon.find_all(class_ = 'store_item'):
                store_id = store.find('strong').string
                store_em = list(store.find('em'))
                store_address = store_em[0].strip()
                store_tel = store_em[4].strip()
                result.append([store_id] + [store_address] + [store_tel])
    return
```

```python
def main():
    result = []
    print('kyochon store crawling >>>>>>>>>>>>>>>>>>>>>>>>>>')
    gyochon_store(result)  
    gyochon_tbl = pd.DataFrame(result, columns=('store','address','phone'))
    gyochon_tbl.to_csv('./kyochon.csv', encoding='utf-8-sig', mode='w', index=True)
    del result[:]
    print(result)
```

```python
if __name__ == '__main__':
     main()

```

# 탐앤탐스 동적 웹 크롤링

```python
from selenium import webdriver
import datetime
import time
```

```python
def TomAndToms_store(result, wd):
    cnt = 1
    for i in range(1,500):        
        wd.get('https://tomntoms.com/store/domestic_store_search.html')
        time.sleep(1)
        try:
            wd.execute_script('select_one(%d)' %i)
            time.sleep(1)
            html2 = wd.page_source
            soup2 = BeautifulSoup(html2,'html.parser')
            tomtom_info = soup2.find(class_='pop-content');
            tomtom = tomtom_info.select('div > .in-bx > .desc')

            tomtom_name = tomtom_info.h2.string
            tomtom_address = tomtom[0].string        
            tomtom_tel = tomtom[1].string

            print('%d 번째 지점' %cnt)
            cnt += 1
            result.append([tomtom_name]+[tomtom_address]+[tomtom_tel])        
        except:
            continue

```

```python
def main():
    result = []
    print('Tomandtoms store crawling >>>>>>>>>>>>>>>>>>>>>>>>>>')
    wd = webdriver.Chrome('./chromedriver.exe')
    TomAndToms_store(result, wd)
    print('Tomandtoms store crawling End >>>>>>>>>>>>>>>>>>>>>>>>>>')
    CB_tbl = pd.DataFrame(result, columns=('store', 'address','phone'))
    CB_tbl.to_csv('./TomAndToms.csv', encoding='utf-8-sig', mode='w', index=True)

if __name__ == '__main__':
     main()
```
